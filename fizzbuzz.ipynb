```python
def fizzBuzz1(n = 100):
    # n: the number to count to (optional) 
    
    for i in range(1, n):

        output = ""

        if not i % 3: output += "Fizz"
        if not i % 5: output += "Buzz"
        
        # if i is not divisible by any of the specified numbers, output will be i
        if output=="":  output = str(i)

        print(output)



def fizzBuzz2(n = 100, divisibilityMap = {3:"Fizz", 5:"Buzz"}):
    # n: the number to count to (optional) 
    # divisibilityMap: a dictionary that specifies divisors and the replacement words (optional)

    for i in range(1, n):

        # in the end, output will be either 'Fizz', 'Buzz', 'FizzBuzz' or str(i)
        output = "" 

        # iterate over the divisibilityMap
        # if i is divisible by the "divisor", add "response" to the string that replaces i
        for divisor, response in divisibilityMap.items():
            if not i % divisor: 
                output += response
        
        # if i is not divisible by any of the specified numbers, output will be i
        if output == "":
            output += str(i)        
    
        print(output)     
```

```python
# fizzBuzz1(100)
fizzBuzz1()
```

```python
fizzBuzz2()
```

```python
fizzBuzz2(100, {2:"two-", 3:"three-", 7:"seven-"})
```
